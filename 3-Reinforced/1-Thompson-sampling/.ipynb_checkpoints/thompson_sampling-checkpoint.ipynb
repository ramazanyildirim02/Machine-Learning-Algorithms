```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
data = pd.read_csv('Ads_CTR_Optimisation.csv')
data
```

```python
N = 10000 # 10000 tıklama
d = 10  # toplam 10 ilan var.
toplam = 0 
secilenler = []
birler = [0] * d
sifirlar = [0] * d
```

```python
for n in range(1,N):
    ad = 0 #seçilen ilan
    max_th = 0
    for i in range(0,d):
        rasbeta = random.betavariate( birler[i] + 1 , sifirlar[i] +1)
        if rasbeta > max_th:
            max_th = rasbeta
            ad = i
    secilenler.append(ad)
    odul = data.values[n,ad] # verilerdeki n. satır = 1 ise ödül 1
    if odul == 1:
        birler[ad] = birler[ad]+1
    else :
        sifirlar[ad] = sifirlar[ad] + 1
    toplam = toplam + odul
print('Toplam Ödül:')   
print(toplam)
```

```python
plt.hist(secilenler)
plt.show()
```
