```python
from sklearn.impute import SimpleImputer
```

```python
b = eksikveri.iloc[:,1:4].values
b
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy= "mean")

Yas = eksikveri.iloc[:,1:4].values

imputer = imputer.fit(Yas)
Yas = imputer.transform(Yas)

```

Kategorik verileri numerik değere çevirme
encoder : nominal,ordinal -> numeric

```python
ulke = eksikveri.iloc[:,0].values
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
ulke = le.fit_transform(eksikveri.iloc[:,0])

ohe = preprocessing.OneHotEncoder()
ulke = ohe.fit_transform(ulke.reshape(-1,1)).toarray()
ulke
```

Parçalanan dataframe leri tek bir dataframe de birleştirme

```python
df1 = pd.DataFrame(data= ulke, index = range(22), columns= ["fr","tr","us"])
```

```python
df2 = pd.DataFrame(data= Yas, index = range(22), columns= ["boy","kilo","yas"])
```

```python
df3 = eksikveri[["cinsiyet"]]
```

```python
sonuc = pd.concat([df1,df2,df3],axis=1)
sonuc
```
