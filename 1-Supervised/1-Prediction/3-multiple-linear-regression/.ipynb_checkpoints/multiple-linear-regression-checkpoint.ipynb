```python
#Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataFrame = pd.read_csv("veriler.csv")
```

```python
dataFrame
```

```python
cinsiyet = dataFrame.iloc[:,-1].values
ulke = dataFrame.iloc[:,0].values
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
ohe = preprocessing.OneHotEncoder()


cinsiyet = le.fit_transform(dataFrame.iloc[:,-1])
ulke = le.fit_transform(dataFrame.iloc[:,0])

cinsiyet = ohe.fit_transform(cinsiyet.reshape(-1,1)).toarray()
ulke = ohe.fit_transform(ulke.reshape(-1,1)).toarray()
```

```python
cinsiyet = pd.DataFrame(data= cinsiyet,index= range(len(cinsiyet)),columns= ["cinsiyet"," "])
ulke = pd.DataFrame(data= ulke,index= range(len(ulke)),columns= ["fr","tr","us"])
```

```python
sonuc = pd.concat([ulke,dataFrame.iloc[:,1:4],cinsiyet.iloc[:,0]],axis=1)
sonuc
```

# Cinsiyet tahmin etme

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( sonuc.drop("cinsiyet",axis=1), sonuc[["cinsiyet"]], test_size=0.33, random_state=42)
```

Model inşası

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)

tahmin = regression.predict(X_test)
tahmin
```

# Boy tahmin etme

```python
boy =sonuc.iloc[:,3:4].values
sol = sonuc.iloc[:,:3]
sag = sonuc.iloc[:,4:]

veri = pd.concat([sol,sag],axis=1)

x_train, x_test,y_train,y_test = train_test_split(veri,boy,test_size=0.33, random_state=0)
```

```python
regression2 = LinearRegression()
regression2.fit(X_train,y_train)

tahmin2 = regression2.predict(X_test)
tahmin2
```

Tüm columnları modele verdiğimizde çok da sağlıklı bir model oluşmayabilir. Peki hangi columnları(değişkenleri) modele verelim? Bunun için BACKWARD ELIMINATION kullanıyoruz.

```python
import statsmodels.api as sm
```

```python
# 1
X = np.append(arr = np.ones((22,1)).astype(int), values=veri, axis=1)

X = X[:,[0,1,2,3,4,5,6]]
model = sm.OLS(boy,X).fit()
print(model.summary())
```

```python
# 2
X = np.append(arr = np.ones((22,1)).astype(int), values=veri, axis=1)

X_l = veri.iloc[:,[0,1,2,3,4,5]].values
X_l = np.array(X_l,dtype=float)
model = sm.OLS(boy,X_l).fit()
print(model.summary())
```

1 ve 2 olarak gösterdiğim kod bloklarının işlevi aynıdır. Dilediğinizi kullanabilirsiniz.

Yukarıdaki rapora bakıldığında:
5.sütunun p-value değeri yüksek olduğu için modelden çıkarabiliriz.


Şimdi 5.sütunu çıkarıp tekrardan teste sokalım

```python
X = np.append(arr = np.ones((22,1)).astype(int), values=veri, axis=1)

X_l = veri.iloc[:,[0,1,2,3,5]].values
X_l = np.array(X_l,dtype=float)
model = sm.OLS(boy,X_l).fit()
print(model.summary())
```

P>|t| kısmına bakınca eldeki sonuçların fena olmadığını ve sadece 5.sütunu çıkararak modeli tekrardan oluşturabileceğimizi düşünüyorum.
