```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("musteriler.csv")
data
```

```python
X = data.iloc[:,3:].values
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, init= "k-means++")
kmeans.fit(X)
Y_tahmin= kmeans.predict(X)
print(Y_tahmin)  
plt.scatter(X[Y_tahmin==0,0],X[Y_tahmin==0,1],s=100, c='red')
plt.scatter(X[Y_tahmin==1,0],X[Y_tahmin==1,1],s=100, c='blue')
plt.scatter(X[Y_tahmin==2,0],X[Y_tahmin==2,1],s=100, c='green')
plt.show()

print(kmeans.cluster_centers_)
```

Burada Ã¶rnek olarak 3 kÃ¼meye ayÄ±rdÄ±k fakat algoritmanÄ±n en iyi bir ÅŸekilde Ã§alÄ±ÅŸmasÄ± iÃ§in kaÃ§ kÃ¼meye ayÄ±rmalÄ±?

```python
results = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init="k-means++", random_state=0)
    kmeans.fit(X)
    results.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),results)
```
